Import libraries

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
```

Read and Load data

```python
ins  = pd.read_csv("insurance.csv")
print(ins)
```

plot age vs charges 

```python
sns.scatterplot(x='age', y = 'charges', data = ins )
plt.title("Age vs Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

Linear regression model and plot of actual vs predicted

```python
X=ins[["age"]]

y = ins["charges"]
# Fit the linear regression model to the data 
simple_model = LinearRegression()
simple_model.fit(X, y) # Predict target values based on input data 
y_pred = simple_model.predict(X)

# Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=ins, x='age', y='charges', label="Data")
plt.plot(X, y_pred, color="red", label="Regression Line")  # regression line in red
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Age vs Charges with Regression Line")
plt.legend()
plt.show()
plt.close()
r2 = r2_score(y, y_pred)
print(f"R Score: {r2:.4f}")
```

Charges increase with age as can be seen from the plot above.
However, the regression line has data points which are far way from the predicted values, which indicates that charges are not only dependent on age, and there could be other 
factors such as gender, smoking status, region, number of children and bmi for example.
The R^2 value of 0.0894 indicates that other factors might play a bigger role than age.
